```python
import nltk
from nltk import RegexpTokenizer
from nltk.util import ngrams
from collections import Counter
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
```

```python
# If you want to bi-gram along with there count the uncomment the bigram count line and return the bigram_counts insted of flattened_bigrams.

def generate_bigrams(text_column):
    
    word_token = RegexpTokenizer(r'\w+')
    lemmatizer = WordNetLemmatizer()

    stop_words = set(stopwords.words('english'))

    tokenized_texts = [word_token.tokenize(text.lower()) for text in text_column]
    lemmatized_texts = [[lemmatizer.lemmatize(word) for word in text if word not in stop_words] for text in tokenized_texts]
   
    bigrams = [list(ngrams(text, 2)) for text in lemmatized_texts]
   
    flattened_bigrams = [bigram for sublist in bigrams for bigram in sublist]

    #bigram_counts = Counter(flattened_bigrams)

    return flattened_bigrams



# Use below function if you return the bigram_counts from generate_bigrams() function

def bigram_to_dataframe(bigram,colnames):
    return pd.DataFrame([(edge[0],edge[1],weight) for edge,weight in bigram.items()],columns=colnames)



# Use below fuction if you return the flattened_bigrams from generate_bigrams() function

def bigram_to_dataframe(bigram,colnames):
    return pd.DataFrame([(edge[0],edge[1]) for edge in bigram],columns=colnames)
```

```python
#Video Tanscript
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt

text = " ".join(i for i in tiktok.video_transcription_text)
stopwords = set(STOPWORDS)
wordcloud = WordCloud(stopwords=stopwords, background_color="white", max_font_size=80, width=800, height=800, max_words=100).generate(text)
plt.figure(figsize=(10,10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
#Positive, Negative, and Neutral
nltk.download('vader_lexicon')
sentiments = SentimentIntensityAnalyzer()
tiktok["Positive"] = [sentiments.polarity_scores(i)["pos"] for i in tiktok["video_transcription_text"]]
tiktok["Negative"] = [sentiments.polarity_scores(i)["neg"] for i in tiktok["video_transcription_text"]]
tiktok["Neutral"] = [sentiments.polarity_scores(i)["neu"] for i in tiktok["video_transcription_text"]]
data = tiktok[["video_transcription_text", "Positive", "Negative", "Neutral"]]
print(tiktok.head())
```

```python
#Positive Analysis
positive_text =' '.join([i for i in tiktok['video_transcription_text'][tiktok['Positive'] > tiktok["Negative"]]])
stopwords = set(STOPWORDS)
wordcloud = WordCloud(stopwords=stopwords, background_color="white", max_font_size=80, width=800, height=800, max_words=100).generate(positive_text)
plt.figure(figsize=(10, 10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
#Negative Analysis
negative_text = ' '.join([i for i in tiktok['video_transcription_text'][tiktok['Negative'] > tiktok['Positive']]])
stopwords = set(STOPWORDS)
wordcloud = WordCloud(stopwords=stopwords, background_color="white", max_font_size=80, width=800, height=800, max_words=50).generate(negative_text)
plt.figure(figsize=(10, 10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
#What distinguishes claim videos from opinion videos.Â 
# Group data by claim_status and calculate mean for the engagement metrics
grouped_data = tiktok.groupby('claim_status')[['video_like_count', 'video_share_count', 'video_comment_count']].mean(numeric_only=True)

# Create a bar plot for each engagement metric
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Plotting like count
axes[0].bar(grouped_data.index, grouped_data['video_like_count'], color='maroon', label='Like Count')
axes[0].set_xlabel('Claim Status')
axes[0].set_ylabel('Average Count')

# Plotting share count
axes[1].bar(grouped_data.index, grouped_data['video_share_count'], color='orange', label='Share Count')
axes[1].set_xlabel('Claim Status')
axes[1].set_ylabel('Average Count')

# Plotting comment count
axes[2].bar(grouped_data.index, grouped_data['video_comment_count'], color='green', label='Comment Count')
axes[2].set_xlabel('Claim Status')
axes[2].set_ylabel('Average Count')

# Add legend
for ax in axes:
    ax.legend()

# Show the plots
plt.tight_layout()
plt.show()
```
